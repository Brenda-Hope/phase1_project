```python
import pandas as pd
import numpy as np
```

##  DATA LOADING

```python
#load the csv data
df = pd.read_csv("AviationData.csv",encoding='latin1')
df.head()
```

 ## FIND DUPLICATE DATA

```python
df=df.drop_duplicates()
df
```

```python
# find the number of rows and columns 
df.shape
```

### In the cell above we have the number of rows and columns respectively

## Finding sum of the missing values

```python
#check for missing values 
df.isna().sum()
```

```python
#find sum of the duplicate data
df.duplicated().sum()
df
```

### We are droping columns that are unnecessary , it has alot of null values 

```python
#drop columns that aren't necessary
cols_to_drop =['Latitude', 'Longitude', 'Airport.Code', 'Airport.Name', 
                'Aircraft.Category', 'FAR.Description', 'Schedule', 'Air.carrier']
df_cleaned =df.drop(columns=cols_to_drop)
df_cleaned
```

###  Replacing the missing values

```python
#drop rows where 'make' or 'model'are missing
df_cleaned.dropna(subset=['Make','Model'],inplace=True)
df_cleaned
```

### Fill in the numerical missing data

```python
#fill in missing numerical data
injury_col = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']
df_cleaned[injury_col] =df_cleaned[injury_col].fillna(0)
df_cleaned
```

### fill in the missing categorical data  

```python
# fill in the missing categorical data 
catg_col=['Location','Weather.Condition', 'Country', 'Injury.Severity','Purpose.of.flight', 
            'Engine.Type','Publication.Date', 'Broad.phase.of.flight','Aircraft.damage' ]
df_cleaned[catg_col]=df_cleaned[catg_col].fillna('unknown')
df_cleaned
```

### Save data

```python
#save the clean data
df_cleaned.to_csv(r"C:\Users\HP\Documents\PROJECT\AviationData_Cleaned.csv", index=False)
```

### Import necessary libraries for visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

fatalities_per_model = df_cleaned.groupby("Model")["Total.Fatal.Injuries"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,7))
sns.barplot(x=fatalities_per_model.values, y=fatalities_per_model.index, palette="coolwarm")
plt.title("Top 10 Aircraft Models with Most Fatalities")
plt.xlabel("Total Fatal Injuries")
plt.ylabel("Aircraft Model")
plt.show()
```

#### The above data shows the  aircraft models with the most fatalities
